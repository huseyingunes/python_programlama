# Flask Dekoratörlerinin Çalışma Mantığı

Bu notebook, Flask'ta `@app.route` gibi dekoratörlerin nasıl çalıştığını açıklar.

## 1. Dekoratör Nedir?
Bir dekoratör, bir fonksiyona ek özellik katar.

```python
def decorator(func):
    def wrapper():
        print("Fonksiyon çalışmadan önce")
        func()
        print("Fonksiyon çalıştıktan sonra")
    return wrapper

@decorator
def selam():
    print("Merhaba!")

selam()
```

## 2. Flask'ta @app.route Bir Dekoratördür

```python
from flask import Flask
app = Flask(__name__)

@app.route('/hello')
def hello():
    return 'Hello!'
```

## 3. Flask'ın route() Fonksiyonunun Basitleştirilmiş Hali

```python
def route(rule):
    def decorator(f):
        print(f"URL kaydedildi: {rule} -> {f.__name__}")
        return f
    return decorator

@route('/deneme')
def test():
    print("Çalıştı")

test()
```
